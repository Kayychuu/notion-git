#Desafío alura Store

1. Extracción de datos

```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

2. EDA

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.pairplot(tienda, kind="kde")
plt.show()
```

```python
for col in tienda.columns:
  sns.boxplot(tienda[col])
  plt.show()
```

```python
for col in tienda.columns:
  sns.histplot(tienda[col])
  plt.show()
```

3. Análisis de datos

```python
# En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.
precioTotal = tienda["Precio"].sum()
precioTotal2 = tienda2["Precio"].sum()
precioTotal3 = tienda3["Precio"].sum()
precioTotal4 = tienda4["Precio"].sum()
print(f"El precio total de la tienda 1 es: {precioTotal}. \nEl precio total de la tienda 2 es: {precioTotal2}. \nEl precio total de la tienda 3 es: {precioTotal3}. \nEl precio total de la tienda 4 es: {precioTotal4}.")
# En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.
print("\n")
calificacionPromedio = tienda["Calificación"].mean()
calificacionPromedio2 = tienda2["Calificación"].mean()
calificacionPromedio3 = tienda3["Calificación"].mean()
calificacionPromedio4 = tienda4["Calificación"].mean()
print(f"La calificación promedio de la tienda 1 es: {calificacionPromedio}.\nLa calificación promedio de la tienda 2 es: {calificacionPromedio2}.\nLa calificación promedio de la tienda 3 es: {calificacionPromedio3}.\nLa calificación promedio de la tienda 4 es: {calificacionPromedio4}.")
# En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.
print("\n")
masVendido1 = tienda["Producto"].value_counts().idxmax()
menosVendido1 = tienda["Producto"].value_counts().idxmin()
masVendido2 = tienda2["Producto"].value_counts().idxmax()
menosVendido2 = tienda2["Producto"].value_counts().idxmin()
masVendido3 = tienda3["Producto"].value_counts().idxmax()
menosVendido3 = tienda3["Producto"].value_counts().idxmin()
masVendido4 = tienda4["Producto"].value_counts().idxmax()
menosVendido4 = tienda4["Producto"].value_counts().idxmin()
print(f"El producto más vendido de la tienda 1 es: {masVendido1}, el producto menos vendido de la tienda 1 es: {menosVendido1}.\n El producto más vendido de la tienda 2 es: {masVendido2}, el producto menos vendido de la tienda 2 es: {menosVendido2}.\n El producto más vendido de la tienda 3 es: {masVendido3}, el producto menos vendido de la tienda 3 es: {menosVendido3}.\n El producto más vendido de la tienda 4 es: {masVendido4}, el producto menos vendido de la tienda 4 es: {menosVendido4}.\n")
# En este paso, debes calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.
print("\n")
costoEnvio = tienda["Costo de envío"].mean()
costoEnvio2 = tienda2["Costo de envío"].mean()
costoEnvio3 = tienda3["Costo de envío"].mean()
costoEnvio4 = tienda4["Costo de envío"].mean()
print(f"El costo de envio promedio de la tienda 1 es: {costoEnvio}.\nEl costo de envio promedio de la tienda 2 es: {costoEnvio2}.\nEl costo de envio promedio de la tienda 3 es: {costoEnvio3}.\nEl costo de envio promedio de la tienda 4 es: {costoEnvio4}.")
print("\n")
categoriamasvendida = tienda["Categoría del Producto"].value_counts().idxmax()
categoriamenosvendida = tienda["Categoría del Producto"].value_counts().idxmin()
categoriamasvendida2 = tienda2["Categoría del Producto"].value_counts().idxmax()
categoriamenosvendida2 = tienda2["Categoría del Producto"].value_counts().idxmin()
categoriamasvendida3 = tienda3["Categoría del Producto"].value_counts().idxmax()
categoriamenosvendida3 = tienda3["Categoría del Producto"].value_counts().idxmin()
categoriamasvendida4 = tienda4["Categoría del Producto"].value_counts().idxmax()
categoriamenosvendida4 = tienda4["Categoría del Producto"].value_counts().idxmin()
print(f"La categoría más vendida de la tienda 1 es: {categoriamasvendida}, la categoría menos vendida de la tienda 1 es: {categoriamenosvendida}.\n La categoría más vendida de la tienda 2 es: {categoriamasvendida}, la categoría menos vendida de la tienda 2 es: {categoriamenosvendida}.\nLa categoría más vendida de la tienda 3 es: {categoriamasvendida}, la categoría menos vendida de la tienda 3 es: {categoriamenosvendida}.\n La categoría más vendida de la tienda 4 es: {categoriamasvendida}, la categoría menos vendida de la tienda 4 es: {categoriamenosvendida}.")
```

4. Visualización

```python
# Categoria de pago por metodo de pago

plt.subplots(figsize=(8, 8))
df_2dhist = pd.DataFrame({
    x_label: grp['Método de pago'].value_counts()
    for x_label, grp in tienda.groupby('Categoría del Producto')
})
sns.heatmap(df_2dhist, cmap='Pastel1')
plt.xlabel('Categoría del Producto')
_ = plt.ylabel('Método de pago')
```

```python
# Categoría del producto vs costo del producto
figsize = (12, 1.2 * len(tienda['Categoría del Producto'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(tienda, x='Costo de envío', y='Categoría del Producto', inner='box', palette='Pastel2')
sns.despine(top=True, right=True, bottom=True, left=True)
```

```python
# Calificación vs cantidad de cuotas
tienda.plot(kind='scatter', x='Calificación', y='Cantidad de cuotas', s=32, alpha=.8, color="purple")
plt.gca().spines[['top', 'right',]].set_visible(False)
```

# Extra

```python
import folium
from IPython.display import IFrame

center_lat = tienda['lat'].mean()
center_lon = tienda['lon'].mean()
m = folium.Map(location=[center_lat, center_lon], zoom_start=10)
for index, row in tienda.iterrows():
    folium.CircleMarker(
        location=[row['lat'], row['lon']],
        radius=5,
        color='blue',
        fill=True,
        fill_color='blue',
        fill_opacity=0.6,
        tooltip=f"Producto: {row['Producto']}<br>Precio: {row['Precio']}<br>Calificación: {row['Calificación']}"
    ).add_to(m)

map_path = '/content/mapa_tienda1.html'
m.save(map_path)
IFrame(map_path, width=700, height=500)
```
